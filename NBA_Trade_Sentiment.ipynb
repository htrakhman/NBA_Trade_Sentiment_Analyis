# Sentiment Analysis 

```python
import re

# pre-process the tweet text
def clean_tweet(tweet):
    text = tweet.replace('RT','')
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", text).split())

tweets_df['clean_text'] =  tweets_df['text'].apply(lambda x: clean_tweet(x))
```

```python
tweets_df
```

```python
from textblob import TextBlob
def calculate_polarity(text):
  analysis = TextBlob(text)
  polarity = analysis.sentiment.polarity
  return polarity

tweets_df['polarity_score'] =  tweets_df['clean_text'].apply(lambda x: calculate_polarity(x))
tweets_df
```

```python
from textblob import TextBlob
def calculate_subjectivity(text):
  analysis = TextBlob(text)
  subjectivity = analysis.sentiment.subjectivity
  return subjectivity

tweets_df['subjectivity_score'] =  tweets_df['clean_text'].apply(lambda x: calculate_subjectivity(x))
tweets_df
```

```python

# categorizing the polarty scores to create overall sentiment result of each tweet
def determine_sentiment(polarity):
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'

tweets_df['sentiment'] =  tweets_df['polarity_score'].apply(lambda x: determine_sentiment(x))
tweets_df
```
